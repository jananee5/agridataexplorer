```python
#SQL QUERIES
{"1.Year-wise Trend of Rice Production Across States (Top 3)":
"""
SELECT agridetails.year, SUM(rice_production_tons) AS total_rice_production
FROM agridetails
WHERE agridetails.state_name IN('West Bengal', 'Uttar Pradesh', 'Punjab')
GROUP BY agridetails.year
ORDER BY agridetails.year
DESC;
""",
"2.Top 5 Districts by Wheat Yield Increase Over the Last 5 Years":
"""
SELECT agridetails.dist_name, AVG(wheat_yield_kg_ha) AS avg_wheat_yield
FROM agridetails
WHERE agridetails.year IN (
        SELECT DISTINCT year 
        FROM agridetails
        ORDER BY year DESC
        LIMIT 5
    )
GROUP BY agridetails.dist_name
ORDER BY avg_wheat_yield DESC
LIMIT 5;
""",
"3.States with the Highest Growth in Oilseed Production (5-Year Growth Rate)":
"""
WITH OilseedData AS (
    SELECT agridetails.state_name, agridetails.year, 
        SUM(oilseeds_production_tons) AS production,
        LAG(SUM(oilseeds_production_tons)) OVER (PARTITION BY state_name ORDER BY year) AS previous_production
    FROM agridetails
    WHERE year IN (SELECT DISTINCT year
            FROM agridetails
            ORDER BY year DESC
            LIMIT 5)
    GROUP BY state_name, year)
SELECT state_name, ROUND((production - previous_production) / previous_production * 100, 2) AS growth_rate
FROM OilseedData
WHERE previous_production IS NOT NULL
ORDER BY growth_rate DESC
LIMIT 10;

""",
"4.District-wise Correlation Between Area and Production for Major Crops (Rice, Wheat, and Maize)":
"""
SELECT agridetails.dist_name,
       SUM(rice_area_ha) AS total_rice_area,
       SUM(rice_production_tons) AS total_rice_production,
       SUM(wheat_area_ha) AS total_wheat_area,
       SUM(wheat_production_tons) AS total_wheat_production,
       SUM(maize_area_ha) AS total_maize_area,
       SUM(maize_production_tons) AS total_maize_production
FROM agridetails
WHERE agridetails.year = (SELECT MAX(year) FROM agridetails)
GROUP BY agridetails.dist_name
ORDER BY total_rice_area DESC;""",
"5.Yearly Production Growth of Cotton in Top 5 Cotton Producing States":
"""
SELECT agridetails.year, agridetails.state_name, SUM(cotton_production_tons) AS total_cotton_production
FROM agridetails
WHERE agridetails.state_name IN (SELECT state_name FROM agridetails GROUP BY state_name ORDER BY SUM(cotton_production_tons) DESC LIMIT 5)
GROUP BY agridetails.year, agridetails.state_name
ORDER BY agridetails.year DESC;
""",
"6.Districts with the Highest Groundnut Production in 2017":
"""
SELECT agridetails.dist_name, SUM(groundnut_production_tons) AS total_groundnut_production
FROM agridetails
WHERE agridetails.year = 2017
GROUP BY agridetails.dist_name
ORDER BY total_groundnut_production DESC
LIMIT 10;
""",
"7.Annual Average Maize Yield Across All States":
"""
SELECT agridetails.year, AVG(maize_yield_kg_ha) AS average_maize_yield
FROM agridetails
GROUP BY agridetails.year
ORDER BY agridetails.year;
""",
"8.Total Area Cultivated for Oilseeds in Each State":
"""
SELECT agridetails.state_name, SUM(oilseeds_area_ha) AS total_oilseeds_area
FROM agridetails
GROUP BY agridetails.state_name
ORDER BY total_oilseeds_area DESC;
""",
"9.Districts with the Highest Rice Yield":
"""
SELECT agridetails.dist_name, MAX(rice_yield_kg_ha) AS highest_rice_yield
FROM agridetails
GROUP BY agridetails.dist_name
ORDER BY highest_rice_yield DESC
LIMIT 10;
""",
"10.Compare the Production of Wheat and Rice for the Top 5 States Over 10 Years":
"""  
SELECT agridetails.year, agridetails.state_name, SUM(rice_production_tons) AS total_rice_production, SUM(wheat_production_tons) AS total_wheat_production
FROM agridetails
WHERE agridetails.state_name IN (SELECT state_name 
FROM agridetails 
GROUP BY state_name 
ORDER BY SUM(rice_production_tons + wheat_production_tons) DESC LIMIT 5)
GROUP BY agridetails.year, agridetails.state_name
ORDER BY agridetails.year DESC, agridetails.state_name;

"""
}
```
